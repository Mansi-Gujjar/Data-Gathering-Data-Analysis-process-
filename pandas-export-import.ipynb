## to_csv

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
# here we export batsman_runs.csv file in folder earlier this file is not there but after this code it'll be there.

temp_df.to_csv('batsman_runs.csv',index=False)
```

```python
# export batsman_vs_team.csv 

df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum').to_csv('batsman_vs_team.csv')
```

## to_excel

```python
df
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df
```

```python
temp_df.to_excel("Batsman___Runs.xlsx")
```

multiple sheets in excel

```python
temp_df.to_excel("output.xlsx",sheet_name='Batsman_runs')
```

```python
temp_df2 = df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum')
```

```python
with pd.ExcelWriter('output.xlsx') as writer:  
    temp_df.to_excel(writer, sheet_name='Sheet_name_1')
    temp_df2.to_excel(writer, sheet_name='Sheet_name_2')
```
