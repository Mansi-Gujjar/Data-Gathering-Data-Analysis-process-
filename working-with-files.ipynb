# CSV Files
- comma seperated values

1. Importing pandas

```python
import pandas as pd
```

 2. Opening a local csv file

```python
df = pd.read_csv('aug_train.csv')
df
```

 3. Opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

 4. Sep Parameter 
- tab files are csv files but they are seperated by tab so sep='\t'
- sometimes first row become our main name row so to fix that we have to give names column as list and on our observation gives names to each coloumn.

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','rating','votes','genres'])
```

5. Index_col parameter
- Generally pandas makes default s.no or index so to remove that we can make one column as index from data.

```python
pd.read_csv('aug_train.csv',index_col='enrollee_id')
```

 6. Header parameter
 - Sometimes header became first row to avoid that we use header=1

```python
pd.read_csv('test.csv',header=1)
```

 7. use_cols parameter
 - if we want to use some columns then we can avoid rest columns using usecols list and in this we provide the columns we want to use.

```python
pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level'])
```

 8. Squeeze parameters
 - squeeze i.e., dataframe became series i.e., we use one column only

```python
gender = pd.read_csv('aug_train.csv', usecols=['gender'])['gender']
gender
```

9. Skiprows/nrows Parameter
- skiprows list contains rows to skip
- nrows i.e., how many rows you want to use

```python
pd.read_csv('aug_train.csv',skiprows=[0,1,2,3])
```

```python
pd.read_csv('aug_train.csv',nrows=100)
```

10. Encoding parameter
- it tells how to read the text

```python
pd.read_csv('zomato.csv',encoding='latin-1')
```

11. Skip bad lines

```python
pd.read_csv('aug_train.csv', sep=',', encoding='latin-1', on_bad_lines='skip')
```

12. dtypes parameter
- change dtype of column as here target column given as flot [1.0, 0.0] but instead of this we can use [1,0] so for this we have to change dtype float to int.

```python
pd.read_csv('aug_train.csv',dtype={'target':int})
```

13. Handling Dates

-generally date given as object i.e., string in pandas but we want that to be date format

```python
pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date']).info()
```

14. Convertors

- if we want to short name or something extra new then using def function we can convert 

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```

15. na_values parameter
- if we want to make some values as nan

```python
pd.read_csv('aug_train.csv',na_values=['Male',])
```

 16. Loading a huge dataset in chunks
 - if we have big dataset then using chunksize we can divide it into equal parts and then using loop we can work with every chunk together 

```python
dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
for chunks in dfs:
    print(chunks.shape)
```
